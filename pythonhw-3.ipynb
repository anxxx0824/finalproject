```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取試算表資料
df = pd.read_excel('python_hw.xlsx')

# 修改欄位名稱和內容
df = df.rename(columns={'生理性別': 'Gender', '習慣讀書地點': 'Reading place'})
df['Gender'] = df['Gender'].replace({'男': 'Male', '女': 'Female'})
df['Reading place'] = df['Reading place'].replace({'咖啡廳': 'Cafe', '宿舍／家中房間': 'Home', '自習室': 'Study room', '其他': 'Other'})

# 計算每個讀書地點男生和女生的人數
male_counts = df[df['Gender'] == 'Male']['Reading place'].value_counts()
female_counts = df[df['Gender'] == 'Female']['Reading place'].value_counts()

# 繪製圓餅圖
fig, (ax1, ax2) = plt.subplots(1, 2)

# 男生的圓餅圖
colors_male = ['#4682B4', '#B0C4DE', '#9ACD32', '#8470FF']
ax1.pie(male_counts, labels=male_counts.index, autopct='%1.1f%%', colors=colors_male)
ax1.set_title('Male')

# 女生的圓餅圖
colors_female = ['#4682B4', '#B0C4DE', '#8470FF', '#9ACD32']
ax2.pie(female_counts, labels=female_counts.index, autopct='%1.1f%%', colors=colors_female)
ax2.set_title('Female')

# 整個圖表的標題
fig.suptitle('The habit of reading place')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取試算表資料
df = pd.read_excel('python_hw.xlsx')

# 修改欄位名稱和內容
df = df.rename(columns={'系院 ※資管系煩請填工學院': 'department', '讀書時是否會聽音樂 ？': 'Music or Not'})
df['department'] = df['department'].replace({'工學院': 'Enginner', '醫學院': 'Medical', '管學院': 'Manager'})
df['Music or Not'] = df['Music or Not'].replace({'有時聽有時不聽': 'Sometimes', '必聽': 'Always', '完全不會聽': 'Never'})

# 計算每個學院習慣讀書地點的數量
eng_counts = df[df['department'] == 'Enginner']['Music or Not'].value_counts()
man_counts = df[df['department'] == 'Manager']['Music or Not'].value_counts()
med_counts = df[df['department'] == 'Medical']['Music or Not'].value_counts()

# 繪製圖表
labels = ['Always', 'Sometimes', 'Never']
eng_values = [eng_counts.get('Always', 0), eng_counts.get('Sometimes', 0), eng_counts.get('Never', 0)]
man_values = [man_counts.get('Always', 0), man_counts.get('Sometimes', 0), man_counts.get('Never', 0)]
med_values = [med_counts.get('Always', 0), med_counts.get('Sometimes', 0), med_counts.get('Never', 0)]

x = range(len(labels))
width = 0.2

fig, ax = plt.subplots()
ax.bar(x, eng_values, width, label='Engineer')
ax.bar([i + width for i in x], man_values, width, label='Manager')
ax.bar([i + 2 * width for i in x], med_values, width, label='Medical')

ax.set_xlabel('Music or Not')
ax.set_ylabel('Count')
ax.set_title('Music or Not by Department')
ax.set_xticks([i + width for i in x])
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
